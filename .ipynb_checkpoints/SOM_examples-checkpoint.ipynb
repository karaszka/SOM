```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy.stats as stats
from math import exp, log
from mpl_toolkits.mplot3d import Axes3D
from som import *

plt.style.use('ggplot')
mpl.rcParams['font.size'] = 20
mpl.rcParams['axes.facecolor'] = 'white'

random.seed(8)

%matplotlib inline
```

```python
#data

lower, upper = -0.5,2.5
mu, sigma = 1, 0.5
X = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
lower, upper = 0.5, 3.5
mu, sigma = 2, 0.5
Y = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
lower, upper = 0, 2
mu, sigma = 1, 0.2
Z = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)

lower, upper = -0.5,2.5
mu, sigma = 1, 0.5
X2 = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
lower, upper = 0.5, 3.5
mu, sigma = 2, 0.5
Y2 = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
lower, upper = 1, 3
mu, sigma = 2, 0.2
Z2 = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)

data1 = np.array((X.rvs(20000),Y.rvs(20000),Z.rvs(20000))).T
data2 = np.array((X2.rvs(20000),Y2.rvs(20000),Z2.rvs(20000))).T
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data1[:,0], data1[:,1], data1[:,2],s=0.6,alpha=0.3,color='k')
ax.scatter(data2[:,0], data2[:,1], data2[:,2],s=0.6,alpha=0.3, color='b')

plt.title("Toy data");
```

```python
# The permutations make it imposible to learn for the SOM with threshold sometimes
dataset1 = np.random.permutation(data1)
dataset2 = np.random.permutation(data2)
data = np.concatenate((dataset1,dataset2))
```

## SOM with threshold

```python
map1 = onlineSOM_learning(10,3)
for i in range(data.shape[0]):
    map1.update(data[i])
map_1 = map1.reshape()  
```

```python
plt.plot(map1.errors,label='generalization error');
plt.plot(map1.err_condition, label='95 th percentile condition');
plt.legend();
```

```python
fig, axs = plt.subplots(2,5,figsize=(25,12),sharex=True, sharey=True,subplot_kw={'projection':'3d'})
to_print = ['100','500', '5000','10000','15000',
           '20100','20500', '25000','30000','35000']
for i,key in enumerate(to_print[:20]):
    m_ = map1.maps_log[key]
    x,y = np.unravel_index(i, [2,5])
    axs[x,y].scatter(dataset1[:2000,0], dataset1[:2000,1], dataset1[:2000,2],s=0.5,alpha=0.2,color='k')
    axs[x,y].scatter(dataset2[:2000,0], dataset2[:2000,1], dataset2[:2000,2],s=0.5, alpha=0.2,color='b')
    axs[x,y].scatter(m_[:,:,0],m_[:,:,1],m_[:,:,2],color='r')
    axs[x,y].plot_wireframe(m_[:,:,0],m_[:,:,1],m_[:,:,2], alpha='0.5')
    axs[x,y].set_title("t = "+key) 

plt.tight_layout()
```

## Generalization errors

```python
l = len(map1.errors)-100
smoothened_err_som = [np.mean(map1.errors[i-100:i]) for i in range(100,l)]
smoothened_err_ubisom = [np.mean(map2.generalization_errors[i-100:i]) for i in range(100,l)]
smoothened_err_plsom = [np.mean(map3.errors[i-100:i]) for i in range(100,l)]
```

```python
plt.figure(figsize=(10,5))
plt.plot(smoothened_err_ubisom,label='smoothed ubiSOM generalization error',alpha=0.8);
plt.plot(smoothened_err_som,label='smoothed plasticSOM generalization error',alpha=0.8);
plt.plot(smoothened_err_plsom,label='smoothed PLSOM generalization error',alpha=0.8);

plt.legend();
```
