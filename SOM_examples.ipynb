```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy.stats as stats
from math import exp, log
from mpl_toolkits.mplot3d import Axes3D
from som import *

plt.style.use('ggplot')
mpl.rcParams['font.size'] = 20
mpl.rcParams['axes.facecolor'] = 'white'

random.seed(8)

%matplotlib inline
```

```python
#data

lower, upper = -0.5,2.5
mu, sigma = 1, 0.5
X = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
lower, upper = 0.5, 3.5
mu, sigma = 2, 0.5
Y = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
lower, upper = 0, 2
mu, sigma = 1, 0.2
Z = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)

lower, upper = -0.5,2.5
mu, sigma = 1, 0.5
X2 = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
lower, upper = 0.5, 3.5
mu, sigma = 2, 0.5
Y2 = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
lower, upper = 1, 3
mu, sigma = 2, 0.2
Z2 = stats.truncnorm(
    (lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)

data1 = np.array((X.rvs(20000),Y.rvs(20000),Z.rvs(20000))).T
data2 = np.array((X2.rvs(20000),Y2.rvs(20000),Z2.rvs(20000))).T
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data1[:,0], data1[:,1], data1[:,2],s=0.6,alpha=0.3,color='k')
ax.scatter(data2[:,0], data2[:,1], data2[:,2],s=0.6,alpha=0.3, color='b')

plt.title("Toy data");
```
